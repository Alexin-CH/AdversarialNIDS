```python
import kagglehub

# Download latest version
path_unsw_nb15 = kagglehub.dataset_download("mrwellsdavid/unsw-nb15")

print("Path to dataset files:", path_unsw_nb15)
```

```python
# List files in the dataset directory
import os
for root, dirs, files in os.walk(path_unsw_nb15):
    for file in files:
        print(os.path.join(root, file))
```

```python
import pandas as pd

feature_file = "NUSW-NB15_features.csv"

# Load and print feature names
# set decoder to 'latin1' to avoid decoding issues
features_df = pd.read_csv(os.path.join(path_unsw_nb15, feature_file), encoding='latin1')
columns = features_df['Name'].tolist()
print("Feature names:", columns)

files = ["UNSW-NB15_1.csv", "UNSW-NB15_2.csv", "UNSW-NB15_3.csv", "UNSW-NB15_4.csv"]

# Set columns for all data files
dataframes = []
for file in files:
    df = pd.read_csv(os.path.join(path_unsw_nb15, file), encoding='latin1')
    df.columns = columns
    dataframes.append(df)

print("Combined DataFrame shape:", df.shape)
print("Column names:", df.columns.tolist())
```
